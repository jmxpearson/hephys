```python
def norm_by_trial(timetuple, method='division'):
    def norm_by_range(df):
        try:
            baseline = df.xs(slice(*ttup), level='time').mean()
        except:
            baseline = 0
        if method == 'division':
            return df.div(baseline)
        elif method == 'subtraction':
            return df - baseline
    return norm_by_range
```

```python
def tf_then_split(series, tffun, events, Tpre, Tpost, *args, **kwargs):
    tfser = 10 * np.log10(tffun(series, *args, **kwargs))
    chunks, idx = phys.evtsplit(tfser, events, Tpre, Tpost)
    if 'normfun' in kwargs:
        nf = kwargs['normfun']
        chunks = [nf(ck) for ck in chunks]
    dfall = pd.concat([ck.mean(axis=1) for ck in chunks], axis=1)
    dfall.columns = tfser.columns
    return dfall
```

### Single trial comparison

Take a single trial and calculate the time-frequency two ways:
1. Wavelet transform whole series, then truncate
2. Truncate series, then wavelet transform

```python
%%time
freqs = np.exp(np.linspace(np.log(2.5), np.log(50), 50))
sertf = tf.continuous_wavelet(ser, freqs=freqs)
```

```python
slc0 = slice(Tpre + stops[0], Tpost + stops[0])
bl0 = (baseline_interval[0] + stops[0], baseline_interval[1] + stops[0])
cwt_then_trunc = sertf.loc[slc0]
trunc_then_cwt = tf.continuous_wavelet(ser.loc[slc0],freqs=freqs)
```

```python
tf.plot_time_frequency(cwt_then_trunc);
tf.plot_time_frequency(trunc_then_cwt);
tf.plot_time_frequency(cwt_then_trunc - trunc_then_cwt);
```

```python
def basic_nf(timetuple, method='division'):
    def norm_by_range(df):
        baseline = df[slice(*timetuple)].mean()
        if method == 'division':
            return df.div(baseline)
        elif method == 'subtraction':
            return df - baseline
    return norm_by_range
```

```python
nf = basic_nf(bl0, method='subtraction')
ctt = nf(cwt_then_trunc)
ttc = nf(trunc_then_cwt)
tf.plot_time_frequency(ctt);
tf.plot_time_frequency(ttc);
tf.plot_time_frequency(ctt - ttc);
```

```python
def plot_time_frequency(spectrum, **kwargs):
    times = spectrum.index
    freqs = spectrum.columns
    z = spectrum.T
    extent = (times[0], times[-1], freqs[0], freqs[-1])
    if 'interpolation' in kwargs:
        interp = kwargs['interpolation']
    else:
        interp = 'bilinear'
    ax = plt.figure().add_subplot(111)
    im = plt.imshow(z, aspect='auto', extent=extent, 
                    origin='lower', interpolation=interp);
    plt.colorbar();
    if 'clim' in kwargs:
        plt.clim(kwargs['clim'])
        
    num_curr_ticks = len(ax.get_yticklabels())
    new_labs = np.exp(np.linspace(np.log(extent[2]), np.log(extent[3]), 
                                  num_curr_ticks))
    ax.set_yticklabels(np.around(new_labs))
    return plt.gcf()
```

### What about a t-stat map?

```python
dT = Tpost - Tpre
Tpre_big = Tpre - dT
Tpost_big = Tpost + dT
```

```python
tf_big = tf.avg_time_frequency(ser, tf.continuous_wavelet, stops, Tpre_big, Tpost_big, freqs=freqs)[slice(Tpre, Tpost)]
bi = (-1.35, -1.0)
tf_big_normed = tf_big - tf_big[slice(*bi)].mean()
```

```python
from scipy.signal import convolve2d

def std_time_frequency(series, tffun, events, Tpre, Tpost, *args, **kwargs):
    specmats, times, fs = tf._per_event_time_frequency(series, tffun, events, Tpre_big, Tpost_big, *args, **kwargs)
    allspecs = np.dstack(specmats)
    stdspec = np.nanstd(allspecs, axis=2)
    
    #smoother = np.array([[0, 1., 0], [1., 2., 1.], [0, 1., 0]])
    #smoother = np.array([[1, 1., 1], [1., 1., 1.], [1, 1., 1]])
    smoother = np.outer([0.5, 1, 0.5], np.ones((10,)))

    smoother = smoother / np.sum(smoother)

    stdspec = convolve2d(np.nan_to_num(stdspec), smoother, mode='same')
    return pd.DataFrame(stdspec, index=times, columns=fs)
```

```python
tf_big_std = std_time_frequency(ser, tf.continuous_wavelet, stops, Tpre_big, Tpost_big, freqs=freqs)[slice(Tpre, Tpost)]
```

```python
plot_time_frequency(tf_big_std);
```

```python
tf_big_t = tf_big_normed / tf_big_std
plot_time_frequency(tf_big_t, clim=(-1.5, 1.5));
```

```python
# now for pops
tf_pop = tf.avg_time_frequency(ser, tf.continuous_wavelet, pops, Tpre_big, Tpost_big, freqs=freqs)[slice(Tpre, Tpost)]
bi = (-1.35, -1.0)
tf_pop_normed = tf_pop - tf_pop[slice(*bi)].mean()
tf_pop_std = std_time_frequency(ser, tf.continuous_wavelet, pops, Tpre_big, Tpost_big, freqs=freqs)[slice(Tpre, Tpost)]
tf_pop_t = tf_pop_normed / tf_pop_std
plot_time_frequency(tf_pop_t, clim=(-1.5, 1.5));
```
