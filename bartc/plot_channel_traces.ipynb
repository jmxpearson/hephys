# Plot average peri-event per-channel power

Plot power as a function of time surrounding a given event, taken as a median across trials. Also produce plots taking the mean across channels.

```python
import pandas as pd
import physutils
from hephys import dbio
import os
from __future__ import division
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use('ggplot')
```

```python
# which patient, dataset to plot
dtup = 18, 1
```

```python
# set up relevant params
bands = ['delta', 'theta', 'alpha', 'beta']  # which bands to limit to
event = 'stop inflating'  # event to split data around
Tpre = -5  # time before event to grab
Tpre_plot = -1.5  # time before event to plot
Tpost = 1.5  # time after event to grab
smwid = 0.2  # smoothing window (in s)
```

```python
# load data
os.chdir(os.path.expanduser('~/code/hephys/bartc'))
dbname = os.path.expanduser('~/data/bartc/plexdata/bartc.hdf5')
lfp = dbio.fetch_all_such_LFP(dbname, *dtup)
```

```python
# bandpass filter
lfpraw = physutils.LFPset(lfp.copy(), meta=lfp.meta.copy())
if bands:
    lfp = lfp.bandlimit(bands)
```

```python
# decimate to 40 Hz
lfp = lfp.decimate(5)
lfpraw = lfpraw.decimate(5)
```

```python
# instantaneous power
lfp = lfp.instpwr()
lfpraw = lfpraw.instpwr()
```

```python
# censor
lfp = lfp.censor()
lfpraw = lfpraw.censor()
```

```python
# get events
evt = dbio.fetch(dbname, 'events', *dtup)
stops = evt[[event] + ['banked']].dropna()[event]
pops = evt['popped'].dropna()
```

```python
# make peri-stop frame
stop_split = lfp.evtsplit(stops, Tpre, Tpost)
stop_means = stop_split.groupby(level=1).median()
stop_df = physutils.LFPset(stop_means, meta=lfp.meta.copy()).zscore()
stop_df = stop_df.smooth(smwid)
```

```python
# make raw peri-stop frame
raw_split = lfpraw.evtsplit(stops, Tpre, Tpost)
raw_means = raw_split.groupby(level=1).median()
raw_df = physutils.LFPset(raw_means, meta=lfpraw.meta.copy()).zscore()
raw_df = raw_df.smooth(smwid)
```

```python
raw_df.plot(legend=False);
plt.xlim([Tpre_plot, Tpost])
plt.title("Broadband")
if bands:
    for bb in bands:
        to_grab = [col for col in stop_df.columns if bb in col]
        this_df = stop_df.loc[:, to_grab]
        this_df.plot(legend=False)
        plt.title("Band = {}".format(bb))
        plt.xlim([Tpre_plot, Tpost])
```

```python
broadband_series = raw_df.mean(axis=1)
broadband_series.name = 'broadband'
series_list = [broadband_series]

if bands:
    for bb in bands:
        to_grab = [col for col in stop_df.columns if bb in col]
        this_df = stop_df.loc[:, to_grab].mean(axis=1)
        this_df.name = bb
        series_list.append(this_df)
mean_df = pd.concat(series_list, axis=1)
mean_df.plot();
plt.xlim([Tpre_plot, Tpost])
plt.title("Mean power across channels");

# place the legend to the right of the plot
ax = plt.gca();
handles, labels = ax.get_legend_handles_labels()
lgd = ax.legend(handles, labels, loc='upper center', bbox_to_anchor=(1.2,0.75))
```
