# Time-Frequency Analysis: Collapse across channels

Here, we analyze each channel separately, but mean at the level of decibel-scaled power.

To begin, do the usual bookkeeping and imports.

```python
import numpy as np
import pandas as pd
import physutils as phys
import physutils.bootstrap as boot
import physutils.tf as tf
import matplotlib.pyplot as plt
import hephys.dbio as dbio
%matplotlib inline
```

Fetch some LFP data to play with:

```python
import os
dbname = os.path.expanduser('~/data/bartc/plexdata/bartc.hdf5')
dtup = 17, 2
```

```python
lfp = dbio.fetch_all_such_LFP(dbname, *dtup)
```

Now censor lfp and zscore so that channels are comparable:

```python
lfpz = lfp.censor().zscore()
```

Get some example events from a behavioral task, the Balloon Analogue Risk Task (BART):

```python
evt = dbio.fetch_all_such(dbname, 'events', *dtup)
```

Let's compare responses to two sets of events, voluntary stops and unexpected balloon pops:

```python
stops = evt['banked'].dropna()
pops = evt['popped'].dropna()
```

We'll look at a window that begins 1.5 seconds before the event and ends half a second after.

```python
Tpre = -2
Tpost = 0.5
baseline_interval = (-1.5, -1.35)
```

```python
def for_all_chans(df, funcname, *args, **kwargs):
    result_list = []
    for col in df.columns:
        outs = getattr(df, funcname)(col, *args, **kwargs)
        result_list.append(outs)
    return result_list
```

```python
%%time
nf = phys.norm_by_trial(baseline_interval, method='subtraction')
tflist = for_all_chans(lfpz, 'avg_time_frequency',
                       stops, Tpre, Tpost,
                       method='wav', normfun=nf,
                       doplot=False)
```

```python
matlist = [t[0] for t in tflist]
mean_mat = np.dstack(matlist).mean(axis=2)
mean_frame = pd.DataFrame(mean_mat, columns=matlist[0].columns,
                          index=matlist[0].index)
tf.plot_time_frequency(mean_frame);
```

Let's try a contrast plot between two event types:

```python
%%time
nf = phys.norm_by_trial(baseline_interval, method='subtraction')
tflist = for_all_chans(lfpz, 'contrast_time_frequency',
                       [stops, pops], Tpre, Tpost,
                       method='wav', normfun=nf,
                       doplot=False)

matlist = [t[0] for t in tflist]
mean_mat = np.dstack(matlist).mean(axis=2)
mean_frame = pd.DataFrame(mean_mat, columns=matlist[0].columns,
                          index=matlist[0].index)
tf.plot_time_frequency(mean_frame);
```

This is very similar to the result when first taking the average LFP across channels.
