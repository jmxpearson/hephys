But are any clusters in this image statistically significant? 

Our strategy will be to shuffle labels for the two trial types and take any clusters in the image above a threshold that exceed the size of 95% of those in the null distribution.

```python
# Make a dataframe containing trial times and labels
t0 = pd.DataFrame({'time':times0, 'label':0})
t1 = pd.DataFrame({'time':times1, 'label':1})
alltimes = pd.concat([t0, t1])
```

```python
# get time-frequency matrix for each time segment
series = lfp.dataframe[17]
tffun = continuous_wavelet
events = alltimes['time']
spectra, taxis, faxis = per_event_time_frequency(series, tffun, events, Tpre, Tpost, normfun=norm_by_trial(baseline_interval))
```

```python
# set threshold (in dB)
thresh = 3.0
```

```python
def plot_diff_map(img, taxis, faxis):
    ax = plt.figure().add_subplot(111)
    extent = (taxis[0], taxis[-1], faxis[0], faxis[-1])
    im = NonUniformImage(ax, interpolation=None, extent=extent)
    im.set_data(taxis, faxis, img.T)
    ax.set_xlim(extent[0], extent[1])
    ax.set_ylim(extent[2], extent[3])
    ax.images.append(im)
```

Now, do the same thing a whole bunch of times:

```python
%%time
thresh = (3,)
niter = 100
pval = 0.05

if len(thresh) == 1:
    thlo = -thresh[0]
    thhi = thresh[0]
else:
    thlo = thresh[0]
    thhi = thresh[1]

cluster_sizes = []
for ind in np.arange(niter):
    labels = np.random.permutation(alltimes['label'])
    pos = physutils.make_thresholded_diff(spectra, labels, lo=None, hi=thhi)
    neg = physutils.make_thresholded_diff(spectra, labels, lo=thlo, hi=None)

    posclus = physutils.label_clusters(pos)
    negclus = physutils.label_clusters(neg)

    cluster_sizes = np.concatenate([
        cluster_sizes,
        physutils.get_cluster_sizes(posclus), 
        -1 * physutils.get_cluster_sizes(negclus)
        ])

cluster_sizes = np.sort(cluster_sizes)
plo = pval / 2.0
phi = 1 - plo
Nlo = np.floor(len(cluster_sizes) * plo)
Nhi = np.ceil(len(cluster_sizes) * phi)
Clo = cluster_sizes[Nlo]
Chi = cluster_sizes[Nhi]
plt.hist(cluster_sizes, 50, normed=1);
plt.vlines([Clo, Chi], *plt.ylim(), colors='red', linestyles='dashed')
```

Let's use this to plot only significant clusters in the real data:

```python
def select_clusters(arr, cluster_inds):
    out = np.empty((len(cluster_inds),) + arr.shape)
    for ind, cnum in enumerate(cluster_inds):
        out[ind] = arr == cnum
    return np.any(out, 0)
        
```

```python
def threshold_clusters(spectra, labels, lo=None, hi=None, 
                       keeplo=None, keephi=None):
    pos = physutils.make_thresholded_diff(spectra, labels, lo=None, hi=thhi)
    neg = physutils.make_thresholded_diff(spectra, labels, lo=thlo, hi=None)
    
    posclus = physutils.label_clusters(pos)
    negclus = physutils.label_clusters(neg)
    
    szhi = np.bincount(posclus.ravel())
    szlo = -np.bincount(negclus.ravel())
    
    hi_inds = np.flatnonzero(szhi >= keephi) 
    lo_inds = np.flatnonzero(szlo <= keeplo)
    
    # get rid of 0, which is background
    hi_inds = np.setdiff1d(hi_inds, np.array(0))
    lo_inds = np.setdiff1d(lo_inds, np.array(0))
    
    hi_img = select_clusters(posclus, hi_inds)
    lo_img = select_clusters(negclus, lo_inds)
    
    return np.logical_or(hi_img, lo_img)
    
```

```python
truelabels = alltimes['label']
signif = threshold_clusters(spectra, truelabels, lo=thlo, hi=thhi, 
                       keeplo=Clo, keephi=Chi)
plot_diff_map(signif, taxis, faxis)
```

Now retrieve the original contrast image from the spectra (doing this way saves the cost of computing the spectra all over again):

```python
img0 = mean_from_events(np.array(spectra)[truelabels == 0], taxis, faxis)
img1 = mean_from_events(np.array(spectra)[truelabels == 1], taxis, faxis)
contrast = img0 / img1
```

We'll using masking to eliminate all entries that are not significant:

```python
mcontrast = contrast.copy()

mcontrast = mcontrast.mask(~signif)
```

```python
plot_time_frequency(mcontrast);
```
