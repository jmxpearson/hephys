# Time-Frequency Phase Amplitude Coupling

Examine the correlation between phase of one frequency band and amplitude of another (AM coupling).

To begin, do the usual bookkeeping and imports.

```python
import numpy as np
import pandas as pd
import physutils as phys
import physutils.bootstrap as boot
import physutils.tf as tf
import matplotlib.pyplot as plt
import hephys.dbio as dbio
# %config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

Fetch some LFP data to play with:

```python
import os
dbname = os.path.expanduser('~/data/bartc/plexdata/bartc.hdf5')
dtup = 18, 1
```

```python
lfp = dbio.fetch_all_such_LFP(dbname, *dtup)
```

Now censor lfp, zscore to normalize, and average across channels:

```python
# standardize across channels
lfpz = lfp.rzscore()

# mean across channels
lfpm = lfpz.mean(axis=1)

# bandpass filter
bands = ['delta', 'theta', 'alpha', 'beta']
lfpbp = phys.LFPset(lfpm, meta=lfp.meta.copy()).bandlimit(bands)

# relabel by band alone
lfpbp.columns = [nn.split('.')[0] for nn in lfpbp.columns]

# censor
lfpc = lfpbp.censor().decimate(5)
```

Now break this into frequency bands:

# Phase-Amplitude Coupling

Phase-amplitude coupling is the amplitude modulation of signals in one frequency band by another. If we have time series $f$ and $g$ with Hilbert transforms $A_f(t) e^{\phi_f(t)}$ and $A_f(t) e^{\phi_f(t)}$, respectively, the phase-amplitude coupling of $f$ and $g$ at lag $\tau$ is defined by 

$$
\mathrm{PAC} = A_f(t) e^{\phi_g(t + \tau)}
$$



```python
ampband = 'beta'
phaseband = 'theta'
pac = pd.Series(tf.phase_amplitude_coupling(lfpc[ampband], lfpc[phaseband]), index=lfpc.index)
```

```python
plt.hist2d(np.real(pac), np.imag(pac), bins=1000, cmap='Blues');
```
